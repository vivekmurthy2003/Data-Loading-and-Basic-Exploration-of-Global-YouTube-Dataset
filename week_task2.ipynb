```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Univariate Analysis**
  

subscribers :(histogram)

```python
data= pd.read_csv('/content/Global YouTube Statistics.csv', encoding='latin1')
```

```python
print(data.columns)
```

```python
# subscribers
plt.figure(figsize=(5,2))
sns.histplot(data,x='subscribers',kde=True)
plt.show()

#kde gives curve
```

Insights: most channels have lower subscriber counts,while few channels have extermely high subscribers

Views : (hist plot)

```python
# Visualize region wise total orders
data.loc[:,'Country'].value_counts()
```

```python

plt.figure(figsize=(17,2))
sns.histplot(data['video views'],kde=True)
plt.title('Distribution of views')

plt.show()
```

Insight: Views are highly skewed with some very popular channels.

Earnings:(box plot)

```python
plt.figure(figsize=(17,2))
sns.boxplot(x=data['highest_yearly_earnings'])
plt.title('Earnings Distribution ')
plt.show()
```

Insight: Presence of extreme outliers indicates some channels earn significantly more.

channel category count:(count plot)

```python

plt.figure(figsize=(10,3))
sns.countplot(y=data['category'])
plt.title('Channel Category Distribution')
plt.show()
```

Insight: Entertainment and Music categories dominate

country distribution:(countplot)

```python
plt.figure(figsize=(7,12))
sns.countplot(y=data['Country'])
plt.title('Channels by Country')
plt.show()
```

Insight: Most channels are from a few major countries.

**Bivariate Analysis**
  * subscribers vs Views

```python
plt.figure(figsize=(10,3))
sns.scatterplot(x=data['subscribers'],y=data['video views'])
plt.title('Subscribers vs Views')
plt.show()
```

Insight: Strong positive relationship between subscribers and views.

Earnings vs subscribers

```python
plt.figure(figsize=(10,3))
sns.scatterplot(x=data['subscribers'],y=data['highest_yearly_earnings'])
plt.title('Earnings vs Subscribers')
plt.show()
```

Insight: Higher subscribers generally lead to higher earnings

category vs Average Views

```python
plt.figure(figsize=(20,6))
sns.barplot(x=data['category'],y=data['video views'])
plt.title('Category vs Average Views')
plt.show()
```

Insight: Some categories generate more views than others.

country vs number of channels

```python
data['Country'].value_counts().plot(kind='bar')
plt.title("Country vs Number of Channels")
plt.show()
```

Insight: Few countries dominate YouTube content creation.

subscribers vs uploads

```python
plt.figure(figsize=(10,3))
sns.scatterplot(x=data['uploads'],y=data['subscribers'])
plt.title('subscribers vs uploads')
plt.show()
```

Insight: Upload frequency does not always guarantee more subscribers.

**Multivariate Analysis**
  * correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True),annot=True)
plt.title('Correlation matrix')
plt.show()
```

Insight: Subscribers, views, and earnings show strong correlation

pairplot

```python
plt.figure(figsize=(10,3))
sns.pairplot(data[['subscribers','video views','highest_yearly_earnings']])
plt.show()
```

Insight: Clear positive relationships among key metrics.

category-wise subscribers by country

```python
plt.figure(figsize=(10,3))
sns.barplot(x=data['Country'],y=data['subscribers'],hue=data['category'])
plt.xticks(rotation=90)
plt.title('Category-wise Subscribers by Country')
plt.show()
```

Insight: Performance varies by category across countries.

**Overall summary**
  

 * ***Subscriber Distribution:*** Most YouTube channels have lower subscriber counts, while only a few channels have extremely high subscribers (skewed distribution).
* ***Views Pattern:*** Video views are highly concentrated among top channels, showing that popular channels dominate viewership.
* ***Earnings Trend:*** Channels with more subscribers and views generally earn higher yearly earnings.
Strong Relationships: There is a strong positive correlation between subscribers, video views, and earnings.
* ***Category Performance:*** Entertainment and Music categories have more channels, views, and higher earnings compared to other categories.
* ***Country Influence:*** A few countries contribute the majority of top YouTube channels, indicating geographical influence on performance.
* ***Uploads vs Subscribers:*** Higher upload frequency does not always guarantee more subscribers.
* ***Outliers:*** Some channels perform extremely better than others in terms of views, subscribers, and earnings.
* **Overall Insight:** Channel success on YouTube is mainly driven by subscriber count, views, and content category, with strong relationships among these factors.
